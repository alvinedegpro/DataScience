**Whitespace Formatting**

Many languages use curly braces to delimit blocks of code. Python uses indentation,
The pound sign marks the start of a comment. Python itself
ignores the comments, but they're helpful for anyone reading the code.
This makes Python code very readable, but it also means that you have to be very
Be careful with your formatting.

````python
for i in [1, 2, 3, 4, 5]: # start of outer loop
    print(i) # print the current value of i
    for j in [1, 2, 3, 4, 5]: # start of inner loop
        print(j) # print the current value of j
        print(i + j) # print the sum of i and j
    print(i) # print the current value of i after inner loop finishes
print("done looping") # print "done looping" after all loops are finished
````

**Object-Oriented Programming**

Like many languages, Python allows you to define classes that encapsulate data and
the functions that operate on them. We’ll use them sometimes to make our code
cleaner and simpler.

A class contains zero or more member functions. By convention, each takes a first
parameter, self, that refers to the particular class instance.

To define a class, you use the class keyword and a PascalCase name

````python
class Student:
  def __init__(self,name,age):
    self.name = name
    self.age = age
  def introduction(self):
    return f"Hi my name is {self.name} and i am {self.age} years old"
student1 = Student("Will",10)
student1.introduction()
````

We’ll also occasionally create subclasses that inherit some of their functionality from a
parent class

````python
class StudentCollege(Student):
  def __init__(self,name,age,major):
    super().__init__(name,age)
    self.major = major
  def smajor(self):
    return f"My Major is {self.major}"
student2 = StudentCollege("John",15,"Computer Science")
student2.smajor()
````

**zip and Argument Unpacking**

Often we will need to zip two or more iterables together. The zip function transforms
multiple iterables into a single iterable of tuples of corresponding function

````python
list1 = ['a', 'b', 'c']
list2 = [1, 2, 3]
letters_numbers =[x for x in zip(list1 ,list2)]
````

You can also “unzip” a list using a strange trick

````python
letters , numbers = zip(*letters_numbers)
print(letters)
print(numbers)
````

**args and kwargs**

`*args`: This allows you to pass any number of positional arguments to the function.

`**kwargs`: This allows you to pass any number of keyword arguments (arguments preceded by identifiers) to the function

````python
def sumofnumbers(*args):
  return sum(args)
sumofnumbers(1,2,3,5,6,7,8)
````

````python
def personaldetails(**kwargs):
  for key,value in kwargs.items():
    print (f"{key} :{value}")
personaldetails(Name="John",Age=20 ,City="Kiev")
````
